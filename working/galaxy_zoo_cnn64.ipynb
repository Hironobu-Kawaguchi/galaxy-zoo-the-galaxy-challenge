```python
DATA_DIR = '../input/galaxy-zoo-the-galaxy-challenge/'
zippath = '../input/galaxy-zoo-the-galaxy-challenge/images_training_rev1.zip'
ORIG_SHAPE = (424,424)
IMG_SHAPE = (64,64)

def get_train_data(df, shape=IMG_SHAPE):
   
    sel = df.values
    ids = sel[:,0].astype(int).astype(str)
    y_batch = sel[:,1:].astype(np.float32)
    x_batch = []
    z = zipfile.ZipFile(zippath)
    for i in tqdm(ids):
        x = resize(np.asarray(Image.open(io.BytesIO(z.read('images_training_rev1/' + i + '.jpg')))), shape) / 255
        x_batch.append(x.astype(np.float32))
    x_batch = np.array(x_batch)
    z.close()
    return x_batch, y_batch

%time train_x, train_y = get_train_data(df)
```

```python
train_x.shape, train_y.shape
```

```python
np.save('../input/train_y', train_y)
```

```python
np.save('../input/train_x', train_x)
```

```python
from sklearn.model_selection import train_test_split
(x_train, x_test, y_train, y_test) = train_test_split(train_x, train_y, test_size=0.2, random_state=0)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, BatchNormalization, GlobalMaxPooling2D
from tensorflow.keras import backend as K
```

```python
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
model = Sequential()
model.add(Conv2D(512, (3, 3), input_shape=(IMG_SHAPE[0], IMG_SHAPE[1], 3)))
model.add(Conv2D(256, (3, 3)))
#model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3)))
model.add(Conv2D(128, (3, 3)))
#model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (3, 3)))
model.add(Conv2D(128, (3, 3)))
#model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(GlobalMaxPooling2D())


model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(37))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adamax', metrics=[root_mean_squared_error])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
history.history
```

```python
#acc, val_accのプロット
plt.plot(history.history["root_mean_squared_error"], label="train", ls="-", marker="o")
plt.plot(history.history["val_root_mean_squared_error"], label="test", ls="-", marker="x")
plt.ylabel("root_mean_squared_error")
plt.xlabel("epoch")
plt.legend(loc="best")
plt.show()
```

```python
test_loss, test_rmse = model.evaluate(x_test, y_test, verbose=1)
```
